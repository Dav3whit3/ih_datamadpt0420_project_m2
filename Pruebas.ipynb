```python
a = diamonds.groupby('color').count().reset_index()
a
```

```python
fig, axes = plt.subplots(ncols=1, nrows=1, figsize=(8, 4))
axes.barh(a['color'], a['carat'])
plt.tight_layout(5)

axes.set(xlim=[0, 15000],
           title='Diamonds per color',
           ylabel='Color type',
           xlabel='nº of diamonds')

axes.axvline(8000, ls=':')
```
